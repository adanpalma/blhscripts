# Graficas Deemed Default Panama
## Importando Librerias Generales

```python
from pandas import DataFrame
import pandas as pd
#import chart_studio.plotly as py
import plotly.express as px

import cufflinks as cf
%matplotlib inline
from plotly.offline import download_plotlyjs,init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

cf.go_offline()

```

## Lectura del CSV con la data a graficar (deemed defaults hipotecas panama)
 )

```python
df_plt = pd.read_csv('data/Pan_Hipo_DeemDefault.csv')
df_plt.drop('Unnamed: 0',axis=1,inplace=True) #Elimino una columna que se genero al grabar el csv
df_plt.head(5)
```


## Grafica

```python
import plotly.graph_objects as go
#dfa = px.data.stocks()
#px.line(data_frame=dfa,x='date',y=['GOOG','AAPL'],labels={'x':'Date','y':'Prices'}, title="Apple vs Google")

fig = go.Figure()
vintage_1 = df_plt.vintage.unique().min()
vintage_n = df_plt.vintage.unique().max()

for vintages in range(vintage_1,vintage_n+1):
    fig.add_trace((go.Scatter(x=df_plt[df_plt.vintage==vintages].period,y=df_plt[df_plt.vintage==vintages].Hipotecas_deemed_acumulado,mode='lines',name=str(vintages))))


fig.update_layout(title=f'Panama, Mortgages Deemed Defaults {vintage_1} - {vintage_n}',
                  xaxis_title="Periods",
                  yaxis_title="Deemed Default %",
                  autosize=True,
                  showlegend=True,
                  plot_bgcolor='light gray',
                  height = 600,
                  width = 1100,
                  margin=dict(
                      l=100,
                      r=100,
                      b=60,
                      t=100,
                      pad=20),
                  legend=dict(orientation='v',
                              yanchor = 'bottom',
                              y = 0),
                  legend_traceorder='reversed',
                  legend_title_text = 'Vintages',

                  xaxis=dict(showline=True,
                             showgrid=False,
                             showticklabels=True,
                             linecolor='rgb(204,204,204)',
                             linewidth=2,
                             ticks='outside',
                             tickfont=dict(family='Arial',size=9,color='rgb(82,82,82)'),
                             tick0=1,
                             dtick=4,
                             tickangle=90),

                  yaxis=dict(showgrid=True,
                             zeroline=True,
                             showline=True,
                             showticklabels=True))


fig.show()

```
